###### **Transformaci√≥n de formatos de DF y pivoteo: Long- Wide**

* Formatos de Datos: Wide vs Long

**¬øQu√© es Wide?**

- En el formato **wide**, cada variable o categor√≠a aparece como una columna.
- Frecuente en reportes y tablas resumen.

Ejemplo:

| Persona | A√±o_2020 | A√±o_2021 | A√±o_2022 |
|---------|----------|----------|----------|
| Ana     |   10     |    15    |   20     |
| Juan    |   5      |    7     |   10     |


**¬øQu√© es Long?**

- En el formato **long**, los niveles de una variable se codifican en filas.
- Tambien conocido como Tidy:

   Los datos tidy cumplen:
> 1. Cada variable es una columna.
> 2. Cada observaci√≥n es una fila.
> 3. Cada valor est√° en una celda.



Ejemplo Long (tidy):

| Persona | A√±o   | Valor |
|---------|-------|-------|
| Ana     | 2020  | 10    |
| Ana     | 2021  | 15    |
| Ana     | 2022  | 20    |
| Juan    | 2020  | 5     |
| Juan    | 2021  | 7     |
| Juan    | 2022  | 10    |


> **Recomendado para an√°lisis, visualizaci√≥n y modelado.**







**Uso de las funciones `melt()` y  `pivot() `**


Sint√°xis

üîÑ melt() ‚Äî Convertir de Wide a Long

```Python

pd.melt(
        frame,                  # DataFrame original
        id_vars=None,            # Columnas que NO se van a     modificar (se mantienen fijas)
        value_vars=None,         # Columnas que se van a modificar    (opcional, si no se pone son todas las que no est√©n en     id_vars)
        var_name=None,           # Nombre de la nueva columna que     contendr√° los nombres de las columnas columnas
        value_name='value'       # Nombre de la nueva columna que contendr√° los valores
)
```

üîÑ pivot() ‚Äî Convertir de Long a Wide

```Python
df.pivot(
          index=None,    # Columnas que ser√°n las filas       (identificadores)
          columns=None,  # Columna que se va a convertir en nombres       de columnas
          values=None    # Columna que se va a convertir en los valores de las celdas
)
```






````python
import pandas as pd
#DF wide
df_wide = pd.DataFrame({
                        'Persona': ['Ana', 'Juan'],
                        'A√±o_2020': [10, 5],
                        'A√±o_2021': [15, 7],
                        'A√±o_2022': [20, 10]
                         })

print(df_wide)
````

````python
#Convertir de wide a long
df_long = df_wide.melt(id_vars='Persona',
                       var_name='A√±o',
                       value_name='Valor')

# Limpiamos el nombre de las columnas
df_long['A√±o'] = df_long['A√±o'].str.replace('A√±o_', '')

df_long
````

* Ejemplo de long a wide

````python
df_wide2 = df_long.pivot(index='Persona',
                         columns='A√±o',
                         values='Valor'
                        ).reset_index()

print(df_wide2)
````
