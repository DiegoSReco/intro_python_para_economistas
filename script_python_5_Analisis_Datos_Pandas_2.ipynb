###### **Transformaci贸n de formatos de DF y pivoteo: Long- Wide**

* Formatos de Datos: Wide vs Long

**驴Qu茅 es Wide?**

- En el formato **wide**, cada variable o categor铆a aparece como una columna.
- Frecuente en reportes y tablas resumen.

Ejemplo:

| Persona | A帽o_2020 | A帽o_2021 | A帽o_2022 |
|---------|----------|----------|----------|
| Ana     |   10     |    15    |   20     |
| Juan    |   5      |    7     |   10     |


**驴Qu茅 es Long?**

- En el formato **long**, los niveles de una variable se codifican en filas.
- Tambien conocido como Tidy:

   Los datos tidy cumplen:
> 1. Cada variable es una columna.
> 2. Cada observaci贸n es una fila.
> 3. Cada valor est谩 en una celda.



Ejemplo Long (tidy):

| Persona | A帽o   | Valor |
|---------|-------|-------|
| Ana     | 2020  | 10    |
| Ana     | 2021  | 15    |
| Ana     | 2022  | 20    |
| Juan    | 2020  | 5     |
| Juan    | 2021  | 7     |
| Juan    | 2022  | 10    |


> **Recomendado para an谩lisis, visualizaci贸n y modelado.**







**Uso de las funciones `melt()` y  `pivot() `**


Sint谩xis

 melt()  Convertir de Wide a Long

```Python

pd.melt(
        frame,                  # DataFrame original
        id_vars=None,            # Columnas que NO se van a     modificar (se mantienen fijas)
        value_vars=None,         # Columnas que se van a modificar    (opcional, si no se pone son todas las que no est茅n en     id_vars)
        var_name=None,           # Nombre de la nueva columna que     contendr谩 los nombres de las columnas columnas
        value_name='value'       # Nombre de la nueva columna que contendr谩 los valores
)
```

 pivot()  Convertir de Long a Wide

```Python
df.pivot(
          index=None,    # Columnas que ser谩n las filas       (identificadores)
          columns=None,  # Columna que se va a convertir en nombres       de columnas
          values=None    # Columna que se va a convertir en los valores de las celdas
)
```






````python
import pandas as pd
#DF wide
df_wide = pd.DataFrame({
                        'Persona': ['Ana', 'Juan'],
                        'A帽o_2020': [10, 5],
                        'A帽o_2021': [15, 7],
                        'A帽o_2022': [20, 10]
                         })

print(df_wide)
````

````python
#Convertir de wide a long
df_long = df_wide.melt(id_vars='Persona',
                       var_name='A帽o',
                       value_name='Valor')

# Limpiamos el nombre de las columnas
df_long['A帽o'] = df_long['A帽o'].str.replace('A帽o_', '')

df_long
````

* Ejemplo de long a wide

````python
df_wide2 = df_long.pivot(index='Persona',
                         columns='A帽o',
                         values='Valor'
                        ).reset_index()

print(df_wide2)
````

````python
print(df_close)
````

````python
#Quitar Date de 铆ndice
df_close = df_close.reset_index()

#Convertir de wide a long
df_close_long = df_close.melt(id_vars='Date',
                              var_name='Empresa',
                              value_name='Precio')
#Limpiamos nombres de empresas
df_close_long['Empresa'] = df_close_long['Empresa'].str.replace('_Close', '')

df_close_long.head(10)
````

**Ejemplo 2: Convertir de formato long a wide preciones de activos**
