```python
# Ejemplo 1: Recuperar datos econ贸micos del Banco Mundial
df_pob_tot = wb.download( indicator= 'SP.POP.TOTL')
df_pob_tot
```

```python
#Lista de paises en Latam
abv_latam = ["ARG", "BOL", "BRA", "CHL", "COL", "CRI", "CUB", "DOM", "ECU",
             "SLV", "GTM", "HND", "MEX", "NIC", "PAN", "PRY", "PER", "PRI",
             "URY", "VEN"]


#Poblacion total de paises
df_pob_tot_LA = wb.download( indicator= ['SP.POP.TOTL',
                                         'SI.POV.GINI',
                                         'NY.GDP.MKTP.KD'],
                              country= abv_latam ,
                              start= 2010,
                              end= 2023)
#Reset index
df_pob_tot_LA = df_pob_tot_LA.reset_index()
#Cambiar nombres de columnas
df_pob_tot_LA.rename(columns = {'SP.POP.TOTL': 'Poblacion total', 'SI.POV.GINI':'Gini', 'NY.GDP.MKTP.KD':'GDP'})
```

```python
df_pob_tot_LA[ df_pob_tot_LA['country'].isin(['Mexico', 'Argentina', 'Colombia'])  ]
```
