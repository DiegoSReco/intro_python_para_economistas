```python
import pandas_datareader as pdr
#from datetime import datetime
```

```python
import pandas_datareader.wb as wb

indicadores = wb.search('')

indicadores[['id', 'name', 'sourceNote']]
```

```python
# Ejemplo 1: Recuperar datos econ√≥micos del Banco Mundial
df_pob_tot = wb.download( indicator= 'SP.POP.TOTL')
df_pob_tot
```

```python
#Lista de paises en Latam
abv_latam = ["ARG", "BOL", "BRA", "CHL", "COL", "CRI", "CUB", "DOM", "ECU",
             "SLV", "GTM", "HND", "MEX", "NIC", "PAN", "PRY", "PER", "PRI",
             "URY", "VEN"]


#Poblacion total de paises
df_pob_tot_LA = wb.download( indicator= ['SP.POP.TOTL',
                                         'SE.ADT.LITR.ZS',
                                         '1.0.HCount.1.90usd'],
                              country= abv_latam ,
                              start= 2010,
                              end= 2023)

df_pob_tot_LA
```
