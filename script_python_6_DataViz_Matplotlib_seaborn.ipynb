###   **6.1. Gráficos para variables categóricas y de distribución**

Los **gráficos categóricos** permite visualizar variables categóricas o la relación entre variables categóricas y númericas, así como la distribución por cada una de las categorías.

**Utilidad y aplicación**
- Comparar valores entre categorías
- Mostrar distribuciones por grupo
- Analizar frecuencias
- Visualizar variables nominales u ordinales

**Tipos de gráficos para variables categóricas**

- Bar plots (gráficos de barras)
- Count plots (gráficos de conteo)





Ejemplo:

`````python
import os
import tempfile
import pyarrow as pa
from zipfile import ZipFile
from io import BytesIO
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def enoe_load(año, n_trim, tabla):
    url_gn_enoe = "https://www.inegi.org.mx/contenidos/programas/enoe/15ymas/microdatos"
    last_two_digits = str(año)[-2:]

    if (2005 <= año <= 2019) and (n_trim <= 4):
        list_url = [f"{url_gn_enoe}/{año}trim{n_trim}_csv.zip",
                    f"{tabla}t{n_trim}{last_two_digits}.csv"]
    elif año == 2020 and n_trim == 1:
        list_url = [
            f"{url_gn_enoe}/{año}trim{n_trim}_csv.zip",
            f"ENOE_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif año == 2020 and n_trim == 2:
        etoe_link = "https://www.inegi.org.mx/contenidos/investigacion/etoe"
        raise Exception(f"Para este trimestre no existe versión ENOE. Prueba con ETOE 2020: {etoe_link}")

    elif (año == 2020 and n_trim == 3) or ((2021 <= año <= 2022) and (1 <= n_trim <= 4)):
        list_url = [
            f"{url_gn_enoe}/enoe_n_{año}_trim{n_trim}_csv.zip",
            f"ENOEN_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif año == 2020 and n_trim == 4:
        list_url = [
            f"{url_gn_enoe}/enoe_n_{año}_trim{n_trim}_csv.zip",
            f"enoen_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif año >= 2023 and n_trim >= 1:
        list_url = [
            f"{url_gn_enoe}/enoe_{año}_trim{n_trim}_csv.zip",
            f"ENOE_{tabla}T{n_trim}{last_two_digits}.csv"]
    else:
        print("URL incorrecto")
        return None
    url_data = list_url[0]
    modulo = list_url[1]
    #Realizo la consulta
    request_data = requests.get(url_data).content
    #Guardo en zip file
    zip_form = ZipFile(BytesIO(request_data))
    #Cargamos datos
    with zip_form.open(modulo) as file:
         df_enoe = pd.read_csv(file, encoding='latin-1', na_values= [pd.NA, " "] ) # na_values= [pd.NA, " "]

    return df_enoe
`````

`````python
#Cargamos enoe SDEM
df_enoe_sdem = enoe_load(2024, 3 , "SDEM")
`````

`````python
#Filtramos personas con entrevista completa y que son residentes habituales en las viviendas
df_enoe_sdem = df_enoe_sdem[(df_enoe_sdem['r_def'] ==0) & ((df_enoe_sdem['c_res']==1) | (df_enoe_sdem['c_res']==3))]

#Permanecen variables de interés
df_enoe_sdem= df_enoe_sdem[[ 'cd_a', 'r_def',  'ent','con','v_sel','tipo','mes_cal', 'n_hog','h_mud','n_ren',  't_loc_tri', 'fac_tri', 'clase1', 'clase2', 'sex','emp_ppal', 'eda', 'scian', 'salario' ,'hrsocup', 'ingocup']]

#Variable edad a numérico
df_enoe_sdem['eda'] = df_enoe_sdem['eda'].replace(' ', np.nan)
df_enoe_sdem['eda']  = pd.to_numeric(df_enoe_sdem['eda'])

#Filtramos en edad de trabajar
df_enoe_sdem = df_enoe_sdem[  (df_enoe_sdem["eda"] >= 15) & (df_enoe_sdem["eda"] < 99)  ]

#Ocupado
df_enoe_sdem['ocupado']= np.where((df_enoe_sdem['clase1']==1) & (df_enoe_sdem['clase2']==1),1,0)

#Identificat ocupados y Sin ingreso
df_enoe_sdem['not_ing'] =np.where(np.isnan(df_enoe_sdem['ingocup'] ) & (df_enoe_sdem['ocupado']==1) , 1, 0)

#Codifiación SCIAN
scian_descripciones = { 1: "Agricultura, Ganadería, etc",2: "Minería", 3: "Generación y distribución de electricidad, suministro de agua y gas", 4: "Construcción",
                        5: "Industrias manufactureras", 6: "Comercio al por mayor", 7: "Comercio al por menor", 8: "Transportes, correos y almacenamiento",
                        9: "Información en medios masivos", 10: "Servicios financieros y de seguros", 11: "Servicios inmobiliarios de alquiler de bienes",
                        12: "Servicios profesionales, científicos y técnicos",13: "Corporativos", 14: "Servicios de apoyo a los negocios y manejo de desechos",15: "Servicios educativos",
                        16: "Servicios de salud y de asistencia social", 17: "Servicios de esparcimiento culturales y deportivos", 18: "Servicios de hospedaje y preparación de alimentos y bebidas",
                        19: "Otros servicios, excepto actividades gubernamentales", 20: "Actividades gubernamentales y de organismos internacionales",21: "No especificado"}
#Codificamos categorías
df_enoe_sdem['SCIAN_b'] = df_enoe_sdem['scian'].map(scian_descripciones)
#Codificación entidad con diccionario Entidades

entidad_dict = { 1:"AGS", 2: "BC", 3: "BCS", 4: "CAMP", 5: "COAH", 6: "COL", 7: "CHIS", 8: "CHIH",
                  9: "CDMX", 10: "DGO", 11: "GTO", 12: "GRO", 13: "HGO", 14: "JAL", 15: "MEX", 16: "MICH",
                  17: "MOR", 18: "NAY", 19: "NL", 20: "OAX", 21: "PUE", 22: "QRO", 23: "QR", 24: "SLP",
                  25: "SIN", 26: "SON", 27: "TAB", 28: "TAMS", 29: "TLAX", 30: "VER", 31: "YUC", 32: "ZAC"}
#Entidad con nombre abreviados
df_enoe_sdem["entidad_ab"] = df_enoe_sdem["ent"].map( entidad_dict)
#Sexo
df_enoe_sdem["clas_sexo"] = df_enoe_sdem["sex"].map({1: "Hombre", 2: "Mujer"}).fillna(np.nan)
#Identificación de informales
df_enoe_sdem['informal'] = df_enoe_sdem['emp_ppal'].map({1:1, 2:0}).fillna(np.nan)


`````

**Visualización de distribuciones porcentuales por categoría con barras**

Los gráficos de barras son herramientas útiles para visualizar la distribución
porcentual de variables categóricas. Estos gráficos permiten comparar rápidamente  la proporción que representa cada categoría respecto al total, facilitando la  identificación de patrones, tendencias y relaciones entre las diferentes categorías.

En este ejemplo, visualizaremos la participación porcentual de los diferentes

¿ Cómo agregar etiquetas a las barras?

Se realiza a través de patches son objetos gráficos que representan formas en un gráfico.

* En un gráfico de barras, cada barra es un patch de tipo rectángulo 📊

* Estos objetos se encuentran en el eje gráfico (ax) y pueden ser accedidos para modificar sus propiedades, como la posición, tamaño o incluso agregarles etiquetas.

`````python
#Defino tema especial
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
#Tamaño
plt.figure(figsize=(10,6))
#Creación de plot
ax = sns.barplot(data=df_ind, y='SCIAN_b', x='Participación Sector', palette='Blues_d')
# Agregar etiquetas a las barras
for p in ax.patches:
    ax.text(
              p.get_width() + 0.2 ,  # Ajuste horizontal para poner la etiqueta al final de la barra
              p.get_y() + p.get_height() / 2,  # Ajuste vertical al centro de la barra
              f'{p.get_width():.2f}%',  # Formato de la etiqueta
              ha='left', va='center', fontsize=  10)
# Títulos y etiquetas
plt.title('Participación porcentual por sector económico (SCIAN)', fontsize=12)
plt.xlabel('Participación (%)', fontsize=12)
plt.ylabel('Sector económico', fontsize=12)
plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
`````

* `ax.patches` es una lista de los rectángulos (barras) que han sido creados en el gráfico.

* `p.get_width()` te da el valor del ancho de cada barra (es decir, el valor de la variable que estás graficando).

* `p.get_y()` y `p.get_height()` te permiten saber la posición de la barra en el eje Y y su altura (la cantidad de valores que representa).

#### **Strip plots**

¿Qué es un **Stripplot**?

Un **stripplot**es un tipo de visualización que muestra la distribución de datos mostrando cada punto individual en una dimensión (horizontal o vertical). Los puntos se distribuyen aleatoriamente en el eje perpendicular para evitar superposiciones, creando un efecto de "dispersión" o "jitter" que permite ver la densidad de la distribución.

¿Cómo nos puede ayudar un Stripplot frente a un Boxplot?
Ventajas del Stripplot:

- **Visualización de datos individuales**: Muestra cada observación individual, permitiendo identificar patrones, clusters o valores atípicos que podrían perderse en un boxplot.
- **Transparencia en la cantidad de datos**: Revela visualmente el tamaño de la muestra, algo que no es evidente en un boxplot.

- **Visualización de la dispersión real**: Muestra la dispersión real de los datos en lugar de resumirla en cuartiles.



Ca

`````python


# 1. Tamaño del plot
plt.figure(figsize=(12, 6))

# 2. Creamos  (stripplot)
sns.stripplot(y='SCIAN_b',
              x='ingocup_log',
              data=df_ocupados_f,
              order=orden_industrias,
              palette='viridis',
              size=3,
              alpha=0.3,
              jitter = True)

# 3. Títulos del plot
plt.title('Distribución de ingreso laboral por industria', fontsize=14)
plt.ylabel('Industria')
plt.xlabel('log(Ingreso laboral mensual)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 4. Mostramos
plt.tight_layout()
plt.show()
`````

Parámetro jitter:

- `jitter=True`: Aplica una cantidad predeterminada de dispersión aleatoria
- `jitter=False`: Los puntos se alinean exactamente en una línea recta
- `jitter=0.25` (u otro valor decimal): Controla la amplitud de la dispersión

El jitter es especialmente útil cuando:

- Tienes muchos puntos de datos
- Hay valores repetidos que se superpondrían


**Limitaciones del Stripplot (donde el Boxplot es mejor):**

- **Conjuntos de datos grandes**: Con muchos puntos, un stripplot puede volverse sobrecargado y difícil de interpretar.

- **Medidas estadísticas**: No muestra directamente estadísticas resumen (mediana, cuartiles) que el boxplot presenta claramente.

- **Comparación entre grupos**: Para comparar múltiples grupos, los boxplots pueden ser más claros al mostrar diferencias en las medianas y dispersiones.


Combinar Striplots y Boxplots


`````python
# 1. Tamaño del plot
plt.figure(figsize=(12, 6))

# 2. Primero creamos el boxplot
sns.boxplot(y='SCIAN_b', x='ingocup_log', data=df_ocupados_f,
            order=orden_industrias, palette='viridis',
            width=0.6, fliersize=0)

# 3. Añadimos el stripplot encima
sns.stripplot(y='SCIAN_b', x='ingocup_log', data=df_ocupados_f,
              order=orden_industrias, palette='viridis',
              size=2, alpha=0.1, jitter=  True)

# 4. Títulos del plot
plt.title('Distribución de ingreso laboral por industria', fontsize=14)
plt.ylabel('Industria')
plt.xlabel('log(Ingreso laboral mensual)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 5. Mostramos
plt.tight_layout()
plt.show()
`````
