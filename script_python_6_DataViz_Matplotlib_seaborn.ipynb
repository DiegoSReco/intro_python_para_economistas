###   **6.1. Gráficos categóricos**

Los **gráficos categóricos** permite visualizar variables categóricas o la relación entre variables categóricas y númericas, así como la distribución por cada una de las categorías.

**Utilidad y aplicación**
- Comparar valores entre categorías
- Mostrar distribuciones por grupo
- Analizar frecuencias
- Visualizar variables nominales u ordinales

**Gráficos recomendados**
- Bar plots (gráficos de barras)
- Count plots (gráficos de conteo)
- Categorical scatter plots (swarm plots)

`````python
import os
import tempfile
import pyarrow as pa
from zipfile import ZipFile
from io import BytesIO
import requests
import pandas as pd
import numpy as np

def enoe_load(año, n_trim, tabla):
    url_gn_enoe = "https://www.inegi.org.mx/contenidos/programas/enoe/15ymas/microdatos"
    last_two_digits = str(año)[-2:]

    if (2005 <= año <= 2019) and (n_trim <= 4):
        list_url = [f"{url_gn_enoe}/{año}trim{n_trim}_csv.zip",
                    f"{tabla}t{n_trim}{last_two_digits}.csv"]
    elif año == 2020 and n_trim == 1:
        list_url = [
            f"{url_gn_enoe}/{año}trim{n_trim}_csv.zip",
            f"ENOE_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif año == 2020 and n_trim == 2:
        etoe_link = "https://www.inegi.org.mx/contenidos/investigacion/etoe"
        raise Exception(f"Para este trimestre no existe versión ENOE. Prueba con ETOE 2020: {etoe_link}")

    elif (año == 2020 and n_trim == 3) or ((2021 <= año <= 2022) and (1 <= n_trim <= 4)):
        list_url = [
            f"{url_gn_enoe}/enoe_n_{año}_trim{n_trim}_csv.zip",
            f"ENOEN_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif año == 2020 and n_trim == 4:
        list_url = [
            f"{url_gn_enoe}/enoe_n_{año}_trim{n_trim}_csv.zip",
            f"enoen_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif año >= 2023 and n_trim >= 1:
        list_url = [
            f"{url_gn_enoe}/enoe_{año}_trim{n_trim}_csv.zip",
            f"ENOE_{tabla}T{n_trim}{last_two_digits}.csv"]
    else:
        print("URL incorrecto")
        return None
    url_data = list_url[0]
    modulo = list_url[1]
    #Realizo la consulta
    request_data = requests.get(url_data).content
    #Guardo en zip file
    zip_form = ZipFile(BytesIO(request_data))
    #Cargamos datos
    with zip_form.open(modulo) as file:
         df_enoe = pd.read_csv(file, encoding='latin-1', na_values= [pd.NA, " "] ) # na_values= [pd.NA, " "]

    return df_enoe
`````

`````python
#Cargamos enoe SDEM
df_enoe_sdem = enoe_load(2024, 3 , "SDEM")
`````

¿ Cómo agregar etiquetas a las barras?

Se realiza a través de patches son objetos gráficos que representan formas en un gráfico.

* En un gráfico de barras, cada barra es un patch de tipo rectángulo 📊

* Estos objetos se encuentran en el eje gráfico (ax) y pueden ser accedidos para modificar sus propiedades, como la posición, tamaño o incluso agregarles etiquetas.

`````python
# Crear el gráfico

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
plt.figure(figsize=(10,6))
ax = sns.barplot(data=df_ind, y='SCIAN_b', x='Participación Sector', palette='Blues_d')
# Agregar etiquetas a las barras
for p in ax.patches:
    ax.text(
              p.get_width() + 0.2 ,  # Ajuste horizontal para poner la etiqueta al final de la barra
              p.get_y() + p.get_height() / 2,  # Ajuste vertical al centro de la barra
              f'{p.get_width():.2f}%',  # Formato de la etiqueta
              ha='left', va='center', fontsize=  10)
# Títulos y etiquetas
plt.title('Participación porcentual por sector económico (SCIAN)', fontsize=12)
plt.xlabel('Participación (%)', fontsize=12)
plt.ylabel('Sector económico', fontsize=12)
plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
`````

* `ax.patches` es una lista de los rectángulos (barras) que han sido creados en el gráfico.

* `p.get_width()` te da el valor del ancho de cada barra (es decir, el valor de la variable que estás graficando).

* `p.get_y()` y `p.get_height()` te permiten saber la posición de la barra en el eje Y y su altura (la cantidad de valores que representa).

**¿Qué son los bins?**

- En los histogramas, los bins (o intervalos) son los rangos de valores en los que se agrupan los datos para representar su distribución. Un histograma divide el rango de valores posibles de los datos en intervalos de igual tamaño o de tamaño variable, y cuenta cuántos datos caen en cada intervalo.


- Tenemos un conjunto de datos sobre edades (que varían de 0 a 100 años), y decidimos usar 10 bins, cada bin podría abarcar un rango de 10 años (0-10, 10-20, 20-30, etc.).

- La elección del número de bins es importante porque influye en la apariencia del histograma y en la interpretación de los datos.




`````python
for bin in range(10, 100, 10):

       plt.figure(figsize=(12,6))
       sns.histplot(df_ocupados['hrsocup'],
                     bins= bin,
                     color='skyblue',
                     stat='density',
                     linewidth=2)

       # Títulos y etiquetas
       plt.title('Distribución de  horas laborales', fontsize=14)
       plt.xlabel('Participación Sector (%)', fontsize=12)
       plt.ylabel('Densidad', fontsize=12)
       plt.tight_layout()
       plt.show()
`````

**Creación de subgráficos con FacetGrid**
Un FacetGrid en Seaborn es una manera de crear una matriz de gráficos (subgráficos) para visualizar cómo cambia una variable dependiendo de las categorías de otra variable.


Pasos
 1. Crear un FacetGrid: Usamos sns.FacetGrid() para crear un marco de trabajo que puede dividir tus gráficos según una o más variables categóricas (como col o row).

2. Mapear un gráfico a este FacetGrid: Con .map(), puedes aplicar el tipo de gráfico que desees (como sns.histplot(), sns.boxplot(), etc.) a cada subgráfico.



```Pyton

# Crear el FacetGrid basado en una categoría
g = sns.FacetGrid(df, col='columna_categorica', height=5)

# Aplicar un gráfico específico a cada Facet (subgráfico)
g.map(sns.histplot, 'variable_a_graficar', otros argumentos ....)
```

`````python
# Crear el FacetGrid para separar por 'clas_sexo'
g = sns.FacetGrid(df_ocupados, col='clas_sexo', height=6)

# Crear un histograma en cada subgráfico
g.map(sns.histplot, 'hrsocup', bins=20, stat='density', linewidth=2) # kde=True,

# Títulos y etiquetas
g.set_axis_labels('Horas laborales', 'Densidad')
g.set_titles('{col_name}')
plt.tight_layout()
plt.show()
`````
