###   **6.1. Gr치ficos categ칩ricos**

Los **gr치ficos categ칩ricos** permite visualizar variables categ칩ricas o la relaci칩n entre variables categ칩ricas y n칰mericas, as칤 como la distribuci칩n por cada una de las categor칤as.

**Utilidad y aplicaci칩n**
- Comparar valores entre categor칤as
- Mostrar distribuciones por grupo
- Analizar frecuencias
- Visualizar variables nominales u ordinales

**Gr치ficos recomendados**
- Bar plots (gr치ficos de barras)
- Count plots (gr치ficos de conteo)
- Categorical scatter plots (swarm plots)

`````python
import os
import tempfile
import pyarrow as pa
from zipfile import ZipFile
from io import BytesIO
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def enoe_load(a침o, n_trim, tabla):
    url_gn_enoe = "https://www.inegi.org.mx/contenidos/programas/enoe/15ymas/microdatos"
    last_two_digits = str(a침o)[-2:]

    if (2005 <= a침o <= 2019) and (n_trim <= 4):
        list_url = [f"{url_gn_enoe}/{a침o}trim{n_trim}_csv.zip",
                    f"{tabla}t{n_trim}{last_two_digits}.csv"]
    elif a침o == 2020 and n_trim == 1:
        list_url = [
            f"{url_gn_enoe}/{a침o}trim{n_trim}_csv.zip",
            f"ENOE_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif a침o == 2020 and n_trim == 2:
        etoe_link = "https://www.inegi.org.mx/contenidos/investigacion/etoe"
        raise Exception(f"Para este trimestre no existe versi칩n ENOE. Prueba con ETOE 2020: {etoe_link}")

    elif (a침o == 2020 and n_trim == 3) or ((2021 <= a침o <= 2022) and (1 <= n_trim <= 4)):
        list_url = [
            f"{url_gn_enoe}/enoe_n_{a침o}_trim{n_trim}_csv.zip",
            f"ENOEN_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif a침o == 2020 and n_trim == 4:
        list_url = [
            f"{url_gn_enoe}/enoe_n_{a침o}_trim{n_trim}_csv.zip",
            f"enoen_{tabla}T{n_trim}{last_two_digits}.csv"]
    elif a침o >= 2023 and n_trim >= 1:
        list_url = [
            f"{url_gn_enoe}/enoe_{a침o}_trim{n_trim}_csv.zip",
            f"ENOE_{tabla}T{n_trim}{last_two_digits}.csv"]
    else:
        print("URL incorrecto")
        return None
    url_data = list_url[0]
    modulo = list_url[1]
    #Realizo la consulta
    request_data = requests.get(url_data).content
    #Guardo en zip file
    zip_form = ZipFile(BytesIO(request_data))
    #Cargamos datos
    with zip_form.open(modulo) as file:
         df_enoe = pd.read_csv(file, encoding='latin-1', na_values= [pd.NA, " "] ) # na_values= [pd.NA, " "]

    return df_enoe
`````

`````python
#Cargamos enoe SDEM
df_enoe_sdem = enoe_load(2024, 3 , "SDEM")
`````

`````python
#Filtramos personas con entrevista completa y que son residentes habituales en las viviendas
df_enoe_sdem = df_enoe_sdem[(df_enoe_sdem['r_def'] ==0) & ((df_enoe_sdem['c_res']==1) | (df_enoe_sdem['c_res']==3))]

#Nos quedamos con variables de inter칠s
df_enoe_sdem= df_enoe_sdem[[ 'cd_a', 'r_def',  'ent','con','v_sel','tipo','mes_cal', 'n_hog','h_mud','n_ren',  't_loc_tri', 'fac_tri', 'clase1', 'clase2', 'sex','emp_ppal', 'eda', 'scian', 'salario' ,'hrsocup', 'ingocup']]
`````

`````python
#Variable edad a num칠rico
df_enoe_sdem['eda'] = df_enoe_sdem['eda'].replace(' ', np.nan)
df_enoe_sdem['eda']  = pd.to_numeric(df_enoe_sdem['eda'])

#Filtramos en edad de trabajar
df_enoe_sdem = df_enoe_sdem[  (df_enoe_sdem["eda"] >= 15) & (df_enoe_sdem["eda"] < 99)  ]

#Ocupado
df_enoe_sdem['ocupado']= np.where((df_enoe_sdem['clase1']==1) & (df_enoe_sdem['clase2']==1),1,0)

#Identificat ocupados y Sin ingreso
df_enoe_sdem['not_ing'] =np.where(np.isnan(df_enoe_sdem['ingocup'] ) & (df_enoe_sdem['ocupado']==1) , 1, 0)

#Codifiaci칩n SCIAN
scian_descripciones = { 1: "Agricultura, Ganader칤a, etc",2: "Miner칤a", 3: "Generaci칩n y distribuci칩n de electricidad, suministro de agua y gas", 4: "Construcci칩n",
                        5: "Industrias manufactureras", 6: "Comercio al por mayor", 7: "Comercio al por menor", 8: "Transportes, correos y almacenamiento",
                        9: "Informaci칩n en medios masivos", 10: "Servicios financieros y de seguros", 11: "Servicios inmobiliarios de alquiler de bienes",
                        12: "Servicios profesionales, cient칤ficos y t칠cnicos",13: "Corporativos", 14: "Servicios de apoyo a los negocios y manejo de desechos",15: "Servicios educativos",
                        16: "Servicios de salud y de asistencia social", 17: "Servicios de esparcimiento culturales y deportivos", 18: "Servicios de hospedaje y preparaci칩n de alimentos y bebidas",
                        19: "Otros servicios, excepto actividades gubernamentales", 20: "Actividades gubernamentales y de organismos internacionales",21: "No especificado"}
#Codificamos categor칤as
df_enoe_sdem['SCIAN_b'] = df_enoe_sdem['scian'].map(scian_descripciones)
#Codificaci칩n entidad con diccionario Entidades

entidad_dict = { 1:"AGS", 2: "BC", 3: "BCS", 4: "CAMP", 5: "COAH", 6: "COL", 7: "CHIS", 8: "CHIH",
                  9: "CDMX", 10: "DGO", 11: "GTO", 12: "GRO", 13: "HGO", 14: "JAL", 15: "MEX", 16: "MICH",
                  17: "MOR", 18: "NAY", 19: "NL", 20: "OAX", 21: "PUE", 22: "QRO", 23: "QR", 24: "SLP",
                  25: "SIN", 26: "SON", 27: "TAB", 28: "TAMS", 29: "TLAX", 30: "VER", 31: "YUC", 32: "ZAC"}
#Entidad con nombre abreviados
df_enoe_sdem["entidad_ab"] = df_enoe_sdem["ent"].map( entidad_dict)
#Sexo
df_enoe_sdem["clas_sexo"] = df_enoe_sdem["sex"].map({1: "Hombre", 2: "Mujer"}).fillna(np.nan)
#Identificaci칩n de informales
df_enoe_sdem['informal'] = df_enoe_sdem['emp_ppal'].map({1:1, 2:0}).fillna(np.nan)



`````

 C칩mo agregar etiquetas a las barras?

Se realiza a trav칠s de patches son objetos gr치ficos que representan formas en un gr치fico.

* En un gr치fico de barras, cada barra es un patch de tipo rect치ngulo 游늵

* Estos objetos se encuentran en el eje gr치fico (ax) y pueden ser accedidos para modificar sus propiedades, como la posici칩n, tama침o o incluso agregarles etiquetas.

`````python
#Defino tema especial
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
#Tama침o
plt.figure(figsize=(10,6))
#Creaci칩n de plot
ax = sns.barplot(data=df_ind, y='SCIAN_b', x='Participaci칩n Sector', palette='Blues_d')
# Agregar etiquetas a las barras
for p in ax.patches:
    ax.text(
              p.get_width() + 0.2 ,  # Ajuste horizontal para poner la etiqueta al final de la barra
              p.get_y() + p.get_height() / 2,  # Ajuste vertical al centro de la barra
              f'{p.get_width():.2f}%',  # Formato de la etiqueta
              ha='left', va='center', fontsize=  10)
# T칤tulos y etiquetas
plt.title('Participaci칩n porcentual por sector econ칩mico (SCIAN)', fontsize=12)
plt.xlabel('Participaci칩n (%)', fontsize=12)
plt.ylabel('Sector econ칩mico', fontsize=12)
plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
`````

* `ax.patches` es una lista de los rect치ngulos (barras) que han sido creados en el gr치fico.

* `p.get_width()` te da el valor del ancho de cada barra (es decir, el valor de la variable que est치s graficando).

* `p.get_y()` y `p.get_height()` te permiten saber la posici칩n de la barra en el eje Y y su altura (la cantidad de valores que representa).

**쯈u칠 son los bins?**

- En los histogramas, los bins (o intervalos) son los rangos de valores en los que se agrupan los datos para representar su distribuci칩n. Un histograma divide el rango de valores posibles de los datos en intervalos de igual tama침o o de tama침o variable, y cuenta cu치ntos datos caen en cada intervalo.


- Tenemos un conjunto de datos sobre edades (que var칤an de 0 a 100 a침os), y decidimos usar 10 bins, cada bin podr칤a abarcar un rango de 10 a침os (0-10, 10-20, 20-30, etc.).

- La elecci칩n del n칰mero de bins es importante porque influye en la apariencia del histograma y en la interpretaci칩n de los datos.




`````python
for bin in range(10, 100, 10):

       plt.figure(figsize=(5,4))
       sns.histplot(df_ocupados['hrsocup'],
                     bins= bin,
                     color='skyblue',
                     stat='density',
                     linewidth=2)

       # T칤tulos y etiquetas
       plt.title('Distribuci칩n de  horas laborales', fontsize=14)
       plt.xlabel('Participaci칩n Sector (%)', fontsize=12)
       plt.ylabel('Densidad', fontsize=12)
       plt.tight_layout()
       plt.show()
`````

**Creaci칩n de subgr치ficos con FacetGrid**
Un FacetGrid en Seaborn es una manera de crear una matriz de gr치ficos (subgr치ficos) para visualizar c칩mo cambia una variable dependiendo de las categor칤as de otra variable.


Pasos
 1. Crear un FacetGrid: Usamos sns.FacetGrid() para crear un marco de trabajo que puede dividir tus gr치ficos seg칰n una o m치s variables categ칩ricas (como col o row).

2. Mapear un gr치fico a este FacetGrid: Con .map(), puedes aplicar el tipo de gr치fico que desees (como sns.histplot(), sns.boxplot(), etc.) a cada subgr치fico.



```Pyton

# Crear el FacetGrid basado en una categor칤a
g = sns.FacetGrid(df, col='columna_categorica', height=5)

# Aplicar un gr치fico espec칤fico a cada Facet (subgr치fico)
g.map(sns.histplot, 'variable_a_graficar', otros argumentos ....)
```

`````python
# Crear el FacetGrid para separar por 'clas_sexo'
g = sns.FacetGrid(df_ocupados, col='clas_sexo', height=6)

# Crear un histograma en cada subgr치fico
g.map(sns.histplot, 'hrsocup', bins=20, stat='density', linewidth=2) # kde=True,

# T칤tulos y etiquetas
g.set_axis_labels('Horas laborales', 'Densidad')
g.set_titles('{col_name}')
plt.tight_layout()
plt.show()
`````
